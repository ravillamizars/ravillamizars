```python
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
```

```python
EmpleadosAttrition=pd.read_csv(r"C:\Users\ricardo.villamizar\OneDrive - American Tower\Documentos Ricardo Villamizar\Script\cursos\entregables\modulo 1\Reto1.csv")
```

```python
Nuevos_EmpleadosAttrition=EmpleadosAttrition.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
Nuevos_EmpleadosAttrition['Year']=pd.to_datetime(Nuevos_EmpleadosAttrition['HiringDate']).dt.year
```

```python
Nuevos_EmpleadosAttrition['YearsAtCompany']=2018-Nuevos_EmpleadosAttrition['Year']
```

```python
Nuevos_EmpleadosAttrition=Nuevos_EmpleadosAttrition.rename(columns={'DistanceFromHome':'DistanceFromHome_km'})
```

```python
Nuevos_EmpleadosAttrition['DistanceFromHome'] = Nuevos_EmpleadosAttrition['DistanceFromHome_km'].str.split(pat=' ').str[0].astype(int)
```

```python
Nuevos_EmpleadosAttrition=Nuevos_EmpleadosAttrition.drop(['DistanceFromHome_km','HiringDate','Year'],axis=1)
```

```python
SueldoPromedioDepto=Nuevos_EmpleadosAttrition.groupby(['Department'])['MonthlyIncome'].mean()
```

```python
SueldoPromedioDepto=SueldoPromedioDepto.reset_index()
```

```python
SueldoPromedioDepto=SueldoPromedioDepto.rename(columns={'MonthlyIncome':'SueldoPromedio'})
```

```python
Nuevos_EmpleadosAttrition_BusinessTravel=pd.get_dummies(Nuevos_EmpleadosAttrition.BusinessTravel, prefix ='BusinessTravel')
Nuevos_EmpleadosAttrition_Department=pd.get_dummies(Nuevos_EmpleadosAttrition.Department, prefix ='Department')
Nuevos_EmpleadosAttrition_EducationField=pd.get_dummies(Nuevos_EmpleadosAttrition.EducationField, prefix ='EducationField')
Nuevos_EmpleadosAttrition_Gender=pd.get_dummies(Nuevos_EmpleadosAttrition.Gender, prefix ='Gender')
Nuevos_EmpleadosAttrition_JobRole=pd.get_dummies(Nuevos_EmpleadosAttrition.JobRole, prefix ='JobRole')
Nuevos_EmpleadosAttrition_MaritalStatus=pd.get_dummies(Nuevos_EmpleadosAttrition.MaritalStatus, prefix ='MaritalStatus')
Nuevos_EmpleadosAttrition_OverTime=pd.get_dummies(Nuevos_EmpleadosAttrition.OverTime, prefix ='OverTime')
Nuevos_EmpleadosAttrition_Attrition=pd.get_dummies(Nuevos_EmpleadosAttrition.Attrition, prefix ='Attrition')
```

```python
Nuevos_EmpleadosAttrition_numerico=pd.concat([Nuevos_EmpleadosAttrition,Nuevos_EmpleadosAttrition_BusinessTravel,Nuevos_EmpleadosAttrition_Department,Nuevos_EmpleadosAttrition_EducationField,Nuevos_EmpleadosAttrition_Gender,Nuevos_EmpleadosAttrition_JobRole,Nuevos_EmpleadosAttrition_MaritalStatus,Nuevos_EmpleadosAttrition_OverTime,Nuevos_EmpleadosAttrition_Attrition],axis=1)
```

```python
Nuevos_EmpleadosAttrition_numerico=Nuevos_EmpleadosAttrition_numerico.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime','Attrition'],axis=1)
```

```python
escalador = preprocessing.MinMaxScaler()
frame_escalado = escalador.fit_transform(Nuevos_EmpleadosAttrition_numerico)
```

```python
Nuevos_EmpleadosAttrition_numerico['MonthlyIncome_norm'] = pd.Series(frame_escalado[:,7])
```
